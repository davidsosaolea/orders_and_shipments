# Librerias

```python
import pandas as pd
from herramientas import obtener_resumen_valores_unicos
```

# Carga

```python
df = pd.read_csv('data/orders_and_shipments.csv')
```

# Exploración

```python
pd.set_option('display.max_columns', None)
df.head(3)
```

```python
# Eliminamos la columna ' Order YearMonth ' por no aportar informacion
df.drop(columns=[' Order YearMonth '], inplace=True)
```

```python
# Combina las tres columnas en una sola columna de tipo string
df['Order date'] = df[' Order Year '].astype(str) + '-' + df[' Order Month '].astype(str).str.zfill(2) + '-' + df[' Order Day '].astype(str).str.zfill(2)
df['Shipping date'] = df['Shipment Year'].astype(str) + '-' + df['Shipment Month'].astype(str).str.zfill(2) + '-' + df['Shipment Day'].astype(str).str.zfill(2)

# Convierte la columna de tipo string a un tipo de datos de fecha
df['Order date'] = pd.to_datetime(df['Order date'], format='%Y-%m-%d')
df['Shipping date'] = pd.to_datetime(df['Shipping date'], format='%Y-%m-%d')

# Ahora, la columna 'Order date' y 'Shipping date' contiene las fechas combinadas en un solo campo.
# Puedes eliminar las columnas originales si ya no las necesitas:
df.drop(columns=[' Order Year ', ' Order Month ', ' Order Day '], inplace=True)
df.drop(columns=['Shipment Year', 'Shipment Month', 'Shipment Day'], inplace=True)
```

```python
# Resta las dos columnas y obtén la diferencia en días
df['Dias de Envio'] = (df['Shipping date'] - df['Order date']).dt.days
```

```python
df.head(3)
```

```python
# visualizamos el nombre de cada columna, tipo de dato, valores unicos y nulos
obtener_resumen_valores_unicos(df)
```

```python
# Exploramos los valores perteneciente a los descuentos 
unique_values = df[' Discount % '].unique()
print(unique_values)
```

```python
# Reemplazar ' - ' por 0 en la columna 'Discount %'
df[' Discount % '] = df[' Discount % '].replace('  -  ', 0)

# Eliminar el signo de porcentaje y convertir la columna ' Discount % ' a tipo float
df[' Discount % '] = df[' Discount % '].str.replace('%', '').astype(float)
```
